# <strong>Games Sales EDA </strong>

This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include

   Rank - Ranking of overall sales

   Name - The games name

   Platform - Platform of the games release (i.e. PC,PS4, etc.)

   Year - Year of the game's release

   Genre - Genre of the game

   Publisher - Publisher of the game

   NA_Sales - Sales in North America (in millions)

   EU_Sales - Sales in Europe (in millions)

   JP_Sales - Sales in Japan (in millions)

   Other_Sales - Sales in the rest of the world (in millions)

   Global_Sales - Total worldwide sales.
   
   Source: Kaggle.com

## <strong> 4) Answering the Questions</strong>

### <Strong> Q1- What is the Sales Growth Rate Over the Years? </Strong>


```python
# Show the sales over the years
df2=df.groupby(['Year'])['Global_Sales'].sum().reset_index()
```

```python
# Display first 4 rows
df2.head(4)
```

```python
# Check 'Year' content
df2['Year'].unique()
```

As shown above there is two years missing (2018, 2019).

```python
# Adding year 2018.
r1={'Year':2018,'Global_Sales':0}
df2 = df2.append(r1, ignore_index=True)
```

```python
# Adding year 2019.
r2={'Year':2019,'Global_Sales':0}
df2 = df2.append(r2, ignore_index=True)
```

```python
# Check of the null values.
df2.isnull().sum()
```

```python
# Check 'Year' content.
df2['Year'].unique()
```

```python
# Sort values by year.
df2=df2.sort_values(by=['Year'])
```

Next step is filling the missing sales value, based on the last 3 years average sales.

```python
# Calculate 3 years average sales.
df2['MA']=df2['Global_Sales'].rolling(3,center=True).mean()
```

```python
# Add average sales to the 'Year' missing value.
df2['Global_Sales'] = np.where(df2['Global_Sales']==0,df2['MA'],df2['Global_Sales'])
```

```python
# Remove 'MA'.
del df2['MA']
```

```python
# Display last 4 rows 
df2.tail(4)
```

After filling the missing values we can calculate sales growth rate.

```python
# Calculate grouth rate over the years.
df2['Sales_Growth'] = df2['Global_Sales'].pct_change()*100
```

```python
df2
```

```python
# Convert 'Year' to Int value type.
df2['Year']=df2['Year'].astype(int)
```

```python
# Plot sales chart.
df2.plot(x='Year',y='Global_Sales',color='mediumseagreen',kind='bar',figsize=(15,14))

plt.show()
```

It's appears that 2008 had more sales.

```python
# Plot growth rate chart
df2.plot(x='Year',y='Sales_Growth',color='mediumseagreen',kind='line',figsize=(14,8))

plt.show()
```

As shown above sales growth jumped to the highest value betwen 2015-2020.

### <Strong> Q2- Which Console Had more Video Games Sales? </Strong>

```python
df3=df.groupby(['Platform'])['Global_Sales'].sum().reset_index()
```

```python
df3
```

```python
# Sort before ploting.
df3=df3.sort_values(by=['Global_Sales'],ascending=False)
```

```python
# Ploting
df3.plot(x='Platform',y='Global_Sales',color='mediumseagreen',kind='bar',figsize=(14,8))

plt.show()
```

PS2 is the king of all consoles, it's might because the long period before launching PS3.
